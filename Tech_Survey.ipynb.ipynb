```python
# Function to do the plot
def plotHbar(y,x,pltitle):
    plt.rcdefaults()
    fig, ax = plt.subplots(figsize=(15,15))
    y_pos = np.arange(len(y))
    ax.barh(y_pos, x, align='center')
    ax.set_yticks(y_pos)
    ax.set_yticklabels(y)
    ax.invert_yaxis()  # labels read top-to-bottom
    ax.set_xlabel('Difference')
    ax.set_title(pltitle)
    plt.show()
```

## How did Bratin's org improve mean score from 2019 to 2020?

```python
df2 = df[df.Statistics=='Mean Score']
xx = df2["sbratin 2020"] - df2["sbratin 2019"]
yy = df2["Question"]
plotHbar(yy,xx,"Mean Score: Bratin 2020 minus Bratin 2019")
```
